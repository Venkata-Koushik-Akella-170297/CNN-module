```python
from keras.datasets import mnist
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
(x_train,_),(x_test,_) = mnist.load_data()
print("shape of x_train dataset is:",x_test.shape)
```

```python
x = x_train[1]
print('the dimension of x is 2D matrix as: ', x.shape)
```

```python
x = cv2.resize(x, (18,18))
plt.imshow(x, cmap='gray')
```

```python
print("range of pixels between 0 and 255")
print("black pixels are more close to 0 and gray pixels are near to 255")
print(x)
```
